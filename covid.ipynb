```python
global_data.data.tail()
#ultima data observada 2020-05-19
```

```python
casos_por_regiao = pd.DataFrame({"Regiao":['Nordeste', 'Sudeste', 'Sul', 'Norte','Centro-Oeste'], "Num_Casos":[0, 0, 0, 0, 0]}) 
print(casos_por_regiao)
acumulado = [0, 0, 0, 0, 0]
for index, row in global_data.iterrows():
    if row['regiao'] != "Brasil" and row['data'] == "2020-05-19":
        if row['regiao'] == 'Nordeste': 
            acumulado[0] += row['casosAcumulado']
            casos_por_regiao.loc[0, 'Num_Casos'] = acumulado[0]   
            
        if row['regiao'] == 'Sudeste':  
            acumulado[1] += row['casosAcumulado']
            casos_por_regiao.loc[1, 'Num_Casos'] = acumulado[1]
                     
        if row['regiao'] == 'Sul': 
            acumulado[2] += row['casosAcumulado']
            casos_por_regiao.loc[2, 'Num_Casos'] = acumulado[2]
                     
        if row['regiao'] == 'Norte':
            acumulado[3] += row['casosAcumulado']
            casos_por_regiao.loc[3, 'Num_Casos'] = acumulado[3]
                    
        if row['regiao'] == 'Centro-Oeste':
            acumulado[4] += row['casosAcumulado']
            casos_por_regiao.loc[4, 'Num_Casos'] = acumulado[4]
```
